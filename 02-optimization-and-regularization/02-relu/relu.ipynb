```python
import numpy as np
import matplotlib.pyplot as plt

def ReLU(x):
    return np.maximum(x,0) # np.maximum function is vectorized, meaning it can operate on entire arrays at once
                           #  also faster than python loop

X = np.linspace(-10, 10)
y_relu = ReLU(X)

plt.plot(y_relu)
```

```python
print("ReLU: ", y_relu)
```

```python
# Leaky ReLU

def LeakyReLU(x):
    return max(x, x*0.1) # using built in max function { not vectorized }

y = [LeakyReLU(i) for i in X]

plt.plot(y)
```

```python
print("ReLU: ", y_relu)
```

```python
print("LeakyReLU: ", y)
```

- We can observe in the output of ReLU negetive values disappear, but not in LeakyReLU
